```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('job_salary_prediction_dataset.csv')
df.head()
```

```python
df.describe()
df.info()
df.isnull().sum()
df['job_title'].value_counts()
print(df.head())
```

```python
sns.histplot(df['salary'], bins=30, kde=True)
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x='experience_years', y='salary', data=df)
plt.title('Salary vs Years of Experience')
plt.xlabel('Internship')
plt.ylabel('Salary')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.pie(df['job_title'].value_counts(), labels=df['job_title'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution of Job Titles')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(df['job_title'].value_counts().index, df['job_title'].value_counts().values)
plt.title('Distribution of Job Titles')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.boxplot(x='job_title', y='salary', data=df)
```

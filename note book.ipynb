```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Excel file

# Convert Income Level to numeric values for plotting
income_map = {"Low": 1, "Medium": 2, "High": 3}
df["Income_Num"] = df["IncomeLevel"].map(income_map)

# Colors for Gender
colors = {"M": "blue", "F": "red"}

plt.figure(figsize=(10,6))

# Scatter plot by gender
for gender in df["Gender"].unique():
    subset = df[df["Gender"] == gender]
    plt.scatter(
        subset["Age"],
        subset["Income_Num"],
        c=colors[gender],
        label=gender,
        alpha=0.6,
        s=50
    )

# Labels
plt.title("Scatter Plot of Age vs Income Level")
plt.xlabel("Age")
plt.ylabel("Income Level")
plt.yticks([1,2,3], ["Low", "Medium", "High"])
plt.legend(title="Gender")
plt.grid(True)

plt.show()
```
